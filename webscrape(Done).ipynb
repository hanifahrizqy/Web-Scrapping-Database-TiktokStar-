```python
import requests
import pandas as pd
```

```python
#The URL that i used in 3/8/2021 14:13 and anythime its changing
topics_url='https://www.famousbirthdays.com/profession/tiktokstar.html'
response=requests.get(topics_url) #output must've in between 200-299
page_content=response.text
```

### Use BeautifulSoup to Parse and Extract Information

```python
!pip install beautifulsoup4 --upgrade --quiet
```

```python
from bs4 import BeautifulSoup
```

```python
doc=BeautifulSoup(page_content,'html.parser')
name_tags=doc.find_all('div',{'class':'name'})
href_tags=doc.find_all('a',{'class':'face person-item clearfix'})
url40=href_tags[40]['href'] #menggunakan herf
rank_tags=doc.find_all('span',{'class':'rank'})
```

```python
#For the Names
titles_artist=[]

for tag in name_tags:
    titles_artist.append(tag.text.strip())

print(titles_artist)
```

```python
#For the URLs
url_artist=[]

for tag in href_tags:
    url_artist.append(tag['href'])

print(url_artist)
```

```python
#still string
rank_artist=[]

for tag in rank_tags:
    rank_artist.append(tag.text.strip())

print(rank_artist)
```

```python
#change into integer
#For the Ranks
for i in range(0, len(rank_artist)):
    rank_artist[i] = int(rank_artist[i])

print(rank_artist)
```

```python
artist_dict={
    'Name':titles_artist,
    'Url':url_artist,
    '*Tiktok Star':rank_artist
}

topic_df=pd.DataFrame(artist_dict)
topic_df[:5]
```

```python
topic_df['Name'] = topic_df['Name'].str.replace('\d+', '')
topic_df['Name'] = topic_df['Name'].str.replace(',', '')
topic_df[:5]
```

```python
topic_df.to_csv('DataAwal2.csv',index=False)
```

### Get Information in each URL

```python
#made the function for each url
def get_dari_awal(topic_url):
    responses=requests.get(topic_url)
    if responses.status_code != 200:
        raise exception ('Failed to load page{}'.format(topic_url))
    doc2=BeautifulSoup(responses.text,'html.parser')
    return doc2

def get_url_repos(doc2):   
    tex_tags=doc2.find_all('p')
    name_text1=doc2.find_all('h1')
    statbox_text1=doc2.find_all('div',{'class':'stat box'})
    popular_text1=doc2.find_all('div',{'class':'btn-rank-wrapper col-xs-6 col-sm-3 col-md-2 col-lg-6'})
    Name=name_text1[0].text.strip()
    real_name=Name.replace('\n\nTikTok Star','')
    About=tex_tags[0].text.strip()
    Before_Fame=tex_tags[1].text.strip()
    Trivia=tex_tags[2].text.strip()
    Family_Life=tex_tags[3].text.strip()
    Associated_with=tex_tags[4].text.strip()
    Birthday=statbox_text1[0].text.strip()
    Birthplace=statbox_text1[1].text.strip()
    Age=statbox_text1[2].text.strip()
    Zodiac_Sign=statbox_text1[3].text.strip()
    Most_Popular=popular_text1[0].text.strip()
    
    column_name_dict={
        'Name':real_name,
        'About':About,
        'Before_Fame':Before_Fame,
        'Trivia':Trivia,
        'Family_Life':Family_Life,
        'Associated_with':Associated_with,
        'Birthday':Birthday,
        'Birthplace':Birthplace,
        'Age':Age,
        'Zodiac_Sign':Zodiac_Sign,
        'Most_Popular':Most_Popular
    }       
    return pd.DataFrame(column_name_dict,index=[0])
```
